```python
'''
Функция для отсеивания признаков с небольшой корреляцией,
написана нейросетью, я ее доработал как смог, но возможны ошибки которые я, в следствие неопытности, не увидел
Я просто не знаю как еще можно уменьшить то безумное количество признаков, если их все запустить, то мало что увидишь
Как уменьшить количество значений в матрице более правильно я не знаю, есть параметр в пандас-профайлинге по порогу корреляции
'''
def correlation_heatmap(df, min_corr=0.8):
    try:
        corr_matrix = df.corr()# Вычисление матрицы корреляций

        # Фильтрация по столбцам с корреляцией min_corr
        selected_features = corr_matrix[(corr_matrix.abs() > min_corr) & (corr_matrix != 1)].dropna(how='all', axis=1).columns.tolist()

        # Отсев столбцов, которые не удовлетворяют условиям фильтрации: 
        df = df[selected_features]
        
        plt.title('Тепловая карта корреляции признаков');

        sns.heatmap(df.corr());

        plt.show();
    except Exception as e:
        print("Error encountered:", e)
```

```python
'''
Выводит процент пропущенных значений в датасете
'''
def nan_percent_plot(df):
    nan_percent = df.isna().mean()*100
    if nan_percent[nan_percent.values > 0].count() > 0:
        (
            nan_percent[nan_percent.values > 0.1] # Вот здесь должен быть ноль, но так как признаков много поставлю поболее
            .sort_values(ascending=True)
            .plot(kind = 'barh',
                  figsize=(16, 12),
                  fontsize=22).set_title('\n' + 'Наибольший процент (относительный) пропущенных значений' + "\n", fontsize=25)
        );
    else:
        print('Пропусков не найдено')
```

Перед тем как заполнять пропуски соседними значениями я принял решение удалить нулевые значение перед заполнением пропусков, так и сделаю:

Надо посмотреть, а много ли абсолютных нулей в признаках, как говорят, что если стоят нули то это значит, либо сбой оборудования, либо они просто не работали и не собирали информацию. Понятное дело, что стандартное отклонение скорее всего изменится, если просто дропнуть строки в которых есть нулевые значения. Я не знаю связано ли оборудование замеряющее признаки или нет, например, произошел сбой в несвязанном оборудовании в этом признаке, но другое оборудование исправно продолжает работать и собирать информацию или же сломалось одно - сломалось все. Изучу нули и посмотрим

```python
'''
Задача в следующем, необходимо оставить только те признаки в которых есть значения равные нулю,
так как при построении барплота график нечитаем из-за большого количества признаков
'''

features_bool = gold_recovery_full.eq(0).sum() != 0 #Получаем булев ряд

columns_isnull = [] #Список для признаков с нулями
for feature_name, feature_bool in zip(gold_recovery_full.columns, features_bool):
    if feature_bool == True:
        columns_isnull.append(feature_name)
```

```python
gold_recovery_full[columns_isnull].eq(0).sum().sort_values().plot(kind='bar', title='Количество строк с нулями', grid=True);
```

Конечно, я не знаю сколько колонок в одной строке содержат нули и сколько строк с наибольшим количеством нулей в одной строке. Принял решение сначала удалить строки равные нулю, а уже затем заполнить пропуски. Распределение имеет ступенчатый характер, скорее всего работа оборудования разделена, сломалось одно оборудование, нули начали появляться в одной группе, сломалось другое - в другой. То есть, я удалю скорее всего несколько тысяч строк, в которых помимо нулей будут другие, возможно полезные признаки, но выбора нет, уйдет очень много времени на обработку нулевых значений, да и не факт, что это даст лучший результат

```python
'''
Эта функция выводит процент изменения чисел до и после заполнения пропусков
обращаем внимание только на наиболее существенные изменения, сами цифры нас не интересуют, важна только динамика и индексы
это функция нейросети, но вроде работает правильно, я до этого не знал как расчитывают процент изменения во фрейме, или забыл
'''
def percentage_change(list1, list2, list3, list4):
    
    # Calculate percentage change and output it in a dataframe
    change1 = ((list2 - list1)/list1) * 100
    change2 = ((list4 - list3)/list3) * 100
    
    output_df = pd.DataFrame()
    output_df['Change 1'] = change1
    output_df['Change 2'] = change2

    return output_df
```

```python
'''
Можно было бы боллее оптимально все это сделать, но время поджимает, как вижу так и пишу
сначала получаем названия колонок, чтобы можно было ориентироваться в массиве numpy (по индексам)
numpy нужен для того чтобы найти разницу до и после заполнения пропусков, вычесть один массив из другого
'''
full_describe = gold_recovery_full.describe() # Замерил основные показатели до заполнения пропусков
full_columns = full_describe.columns

# Взял только два показателя, если надо больше - скажи и какие именно, и почему))
full_std_before = round(full_describe.loc['std'], 3)
full_mean_before = round(full_describe.loc['mean'], 3)
```

```python
gold_recovery_full['date'] = pd.to_datetime(gold_recovery_full['date'], format='%Y-%m-%d %H:%M:%S')

#Так как значения часто близки друг ко другу, если даты рядом, просто сортируем по дате и заполняем соседними значениями
gold_recovery_full = gold_recovery_full.sort_values(by=['date'], ascending=True)
gold_recovery_full.fillna(method='ffill', inplace=True)
```

```python
'''
Эта функция выводит процент изменения чисел до и после заполнения пропусков
обращаем внимание только на наиболее существенные изменения, сами цифры нас не интересуют, важна только динамика и индексы
это функция нейросети, но вроде работает правильно, я до этого не знал как расчитывают процент изменения во фрейме, или забыл
'''
def percentage_change(list1, list2, list3, list4):
    
    # Calculate percentage change and output it in a dataframe
    change1 = ((list2 - list1)/list1) * 100
    change2 = ((list4 - list3)/list3) * 100
    
    output_df = pd.DataFrame()
    output_df['Change 1'] = change1
    output_df['Change 2'] = change2

    return output_df
```

## Модель

```python
# Вероятно из-за этой функции рандомный поиск параметров ооочень медленно работает, но еще виновато большое количество признаков
def sMAPE(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(2 * np.abs(y_pred - y_true) / (np.abs(y_pred) + np.abs(y_true))) * 100

def result_sMAPE(sMAPE_1, sMAPE_2):
    return (0.25 * sMAPE_1) + (0.75 * sMAPE_2)

# Создадим оценщик для рандомсёрча на основе формулы выше, чем ниже значение тем лучше (False)
smape_scorer = make_scorer(sMAPE, greater_is_better=False)
```
